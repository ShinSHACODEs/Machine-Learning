```python
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42) # Indented this line
    kmeans.fit(X_scaled) # Indented this line
    inertia.append(kmeans.inertia_) # Indented this line
# Plot with adjustments to make it visually appealing
plt.figure(figsize=(10, 6))
plt.plot(K, inertia, 'bo-',
markersize=8,
color='teal',
linewidth=2)

plt.title('Elbow Method for Optimal K',

fontsize=16,
fontweight='bold')

plt.xlabel('Number of Clusters (K)',

fontsize=14)
plt.ylabel('Inertia',
fontsize=14)
plt.grid(True, which='both',

linestyle='--', linewidth=0.7)

plt.xticks(K, fontsize=12)
plt.yticks(fontsize=12)
# Show the plot
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X_scaled)
```

```python
data['Cluster'] = kmeans.labels_
centers = kmeans.cluster_centers_
centers_original = scaler.inverse_transform(centers)
```

```python
# Plot the Clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(x='Annual Income (k$)',
y='Spending Score (1-100)',
hue='Cluster',
data=data,
palette='Set2',
s=100,
edgecolor='black',
linewidth=1)

plt.title('Customer Segments Based on Annual Income and Spending Score',
fontsize=16, fontweight='bold')

plt.xlabel('Annual Income (k$)',

fontsize=14)

plt.ylabel('Spending Score (1-100)',

fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True,

linestyle='--',
linewidth=0.7)

# Show the plot
plt.show()
```

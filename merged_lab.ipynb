```python

import pandas as pd
```

```python
movielink  = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/movies_links.csv')
moviemeta  = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/movies_metadata.csv')
movieview  = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/movies_viewer.csv')
```

```python
merged_movies = pd.merge(movielink, movieview, left_on='movieId',right_on='movieId', how='left')
print(merged_movies.head())
```

```python
merged_all = pd.merge(merged_movies, moviemeta, left_on='tmdbId',right_on='id', how='left')
print(merged_all.head())
```

```python
transaction = merged_all.groupby('userId')['original_title'].apply(list).tolist()
```

```python
transaction
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(transaction).transform(transaction)
transaction_df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
transaction_df.head()
```

```python
top_20_columns = transaction_df.sum().sort_values(ascending=False).head(20).index
top_20_columns
```

```python
top_20_df = transaction_df[top_20_columns]
```
